```python
survived = dataset[dataset["Survived"] == 1]
not_survived = dataset[dataset["Survived"] == 0]
```

```python
survived
```

```python
sns.histplot(survived["Age"].dropna(), kde=False, bins=20, label="Survived")
sns.histplot(not_survived["Age"].dropna(), kde=False, bins=20, label="Not Survived")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend()
plt.show()
```
